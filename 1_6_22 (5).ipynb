```python
batch_1["tweet"] = batch_1["tweet"].replace(np.nan, "i")
batch_1 = batch_1[batch_1.tweet != "null"]
batch_1["tweet"] = batch_1["tweet"].replace(0, "i")
batch_1 = batch_1[batch_1.tweet != "i"]
```

```python
tokenized = batch_1["tweet"].apply((lambda x: tokenizer.encode(x, add_special_tokens=True)))
```

```python
import numpy as np
max_len = 0
for i in tokenized.values:
    if len(i) > max_len:
        max_len = len(i)

padded = np.array([i + [0]*(max_len-len(i)) for i in tokenized.values])
```

```python
attention_mask = np.where(padded != 0, 1, 0)
attention_mask.shape
```

```python
input_ids = torch.tensor(padded)  
attention_mask = torch.tensor(attention_mask)
```

```python
input_ids
```

```python
with torch.no_grad():
  last_hidden_states = bertweet(input_ids, attention_mask=attention_mask)
```

```python
features = last_hidden_states[0][:,0,:].numpy()
```

```python
labels = batch_1["ID"]
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels)
```

```python
from sklearn.linear_model import LogisticRegression
lr_clf = LogisticRegression(C=2, max_iter=120)
lr_clf.fit(train_features, train_labels)
y_predL = lr_clf.predict(test_features)
lr=lr_clf.score(test_features, test_labels)
#lr_clf.score(test_features, test_labels)
print('Accuracy score= {:.2f}'.format(lr_clf.score(test_features, test_labels)))
```
